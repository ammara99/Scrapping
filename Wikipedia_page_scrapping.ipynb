```python
import wikipedia as wiki
import pandas as pd
lis =wiki.search("Sports",results=15)
print(lis)
```

```python
title=[]
links=[]
ref=[]
content=[]
summary=[]
for i in lis:
 try:
  page =wiki.page(i)
  title.append( page.original_title)
  links.append( page.links) 
  ref.append(page.references)
  content.append(page.content)
  summary.append(page.summary)
 except:
  continue
```

```python
cleaner =lambda lis: [ x.replace("\n","").replace("=","").replace("]","").replace("[","").replace(")","").replace("(","").replace("'\'","").replace(",","").replace('"',"").replace('\r',"").replace('\b',"").replace('-',"").replace(':',"").replace(';',"") for x in lis]
cleanit=lambda lis:[ cleaner(x) for x in lis]

summary=cleaner(summary)
links=cleanit(links)
content=cleaner(content)
ref=cleanit(ref)
title=cleaner(title)

df=pd.DataFrame({
    "title" : title ,
    "links" : links ,
    "content" : content ,
    "references " : ref ,
    "summary" : summary
})
```
