# importing Libraries

```python
import bs4
from bs4 import BeautifulSoup
from urllib.request import urlopen 
```

# Extracting URL

```python
html=urlopen('https://en.wikipedia.org/wiki/List_of_terrorist_incidents_in_Pakistan_since_2001')
```

# Parsing the html File

```python
bs=BeautifulSoup(html.read(),'html.parser')
```

# Printing page contents

```python
print(bs)
```

# Printing Tags Heading

```python
headers=bs.findAll(['h1','h2','h3','h4','h5','h6'])
print(headers)
```

# Making Lists from Class

```python
namelist=bs.findAll('span',{'class':'mw-headline'})
for name in namelist:
    print(name.get_text())
```

# Printing Text from Class using Loop

```python
for i in bs.select('.mw-headline'):
    print (i.text)
```

# Printing Tags ParaGraphs

```python
bs.select('p')
```

# Printing Text in Paragraph

```python
for i in bs.select('p'):
    print(i.text)
```

# Printing Table 

```python
for child in bs.find('table',{'class':'infobox vevent'}).children:
    print(child)
```

# Printing Table Siblings

```python
for sibling in bs.find('table', {'class':'infobox vevent'}).tr.next_siblings:
    print(sibling)
```
